# Open-CV tutorial

## Open-CV is the computer vision library (reading images, transforms e.t.c)

```python
# Imports
import numpy as np
import cv2
```

```python
img_dir = './Images/img7.jpg' # Path to image (./ indicates everything before the local directory)
img = cv2.imread(img_dir, cv2.IMREAD_COLOR) # COLOR is first read in as BGR
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # Converting colors
#Now we want to view the images
```

# Matplotlib Tutorial

## Matplotlib.pyplot is for viewing data (images, matrices e.t.c)

```python
# Import
import matplotlib.pyplot as plt
```

```python
plt.imshow(img) # Viewing the BGR image
```

```python
# Viewing the RGB image
plt.imshow(img1)
```

```python
# Viewing the images side by side
fig, ax = plt.subplots(1, 2, figsize = (10, 10)) # Indicates the dimesnions of the image array (rows, columns) and size
ax[0].imshow(img)
ax[1].imshow(img1)
```

```python
# We can even read in images using matplotlib
import matplotlib.image as mpimg
img = mpimg.imread(img_dir) # It even goes directly to RGB
plt.imshow(img) 
```

```python
# A sepatate function to show matrices
mat = np.array([[1, 2], [3, 4]])
plt.matshow(mat)
```

```python
# We can plot graphs in pyplot
plt.plot([1, 2, 3, 4, 5])
```

```python
# And add labels
plt.plot([1, 2, 3, 4, 5], 'r') # Change color too
plt.ylabel("y")
plt.xlabel("x")
plt.title("My Graph")
```

```python
# More complex graphs
X = np.linspace(-np.pi, np.pi)
y_sin, y_cos = np.sin(X), np.cos(X)
plt.plot(X, y_sin)
plt.plot(X, y_cos)
plt.ylabel("Y")
plt.xlabel("X")
plt.title("My Graphs")
```

```python
# There is yet another way to read in images, so pick your favorite and stick with that
import skimage.io as io
img = io.imread(img_dir)
plt.imshow(img)
```

# OS, operating system library

## Oftentimes, you need to be able to access local directories

```python
# We have a folder called Images located in the local directory and need to access the images inside
import os
folders = os.listdir('./Images')
folders
```

```python
# To read the images, we need to combine the folder path and the image path
# Let's try this on img1.jpeg

img_dir_name = folders[0]
full_path = os.path.join('./Images', img_dir_name)
print(full_path) # Looks kinda weird but works
```

```python
img = io.imread(full_path)
plt.imshow(img)
```

```python
# Let's get all of the images and create a grid of all of them
fig, ax = plt.subplots(4, 2, figsize = (20, 20))
fig.delaxes(ax[3, 1])
for idx, img_dir in enumerate(os.listdir('./Images')):
    full_dir = os.path.join('./Images', img_dir)
    img1 = io.imread(full_dir)
    img = cv2.resize(img1, (500, 500))
    if idx <= 3:
        ax[idx, 0].imshow(img)
    else:
        ax[idx-4, 1].imshow(img)
```
